```python
#Нарисуйте график функции: y(x) = k∙cos(x – a) + b для некоторых (2-3 различных) значений параметров k, a, b
#Задание (в программе)
#exp(x) + x∙(1 – y) = 1 Решите систему уравнений:
#y = x2 – 1
#Решите систему уравнений и неравенств:
#y = x2 – 1
#exp(x) + x∙(1 – y)  - 1 > 0
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

def f(x, k, a, b):
    return (k*np.cos(x - a) + b)

x = np.linspace(-3*np.pi, 3*np.pi, 201)
plt.plot(x, f(x, 1, 0, 0), color = 'y')
plt.plot(x, f(x, 4, 1, 2), color = 'b')
plt.plot(x, f(x, -1, 2, -1), color = 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#Задание (в программе)
#Решите систему уравнений:
#exp(x) + x∙(1 – y) = 1 
#y = x2 – 1
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,8) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return ( x**2 - 1-y, (np.exp(x) + x - 1)/x -y)

x1, y1 =  fsolve(equations, (-1, 1))
x2, y2 =  fsolve(equations, (2, 4))
              
print (x1, y1)
print (x2, y2)
```
