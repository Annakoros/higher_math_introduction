```python
# Задача 1. Контрольная работа состоит из пяти вопросов. 
#На каждый вопрос приведено четыре варианта ответа, один из которых правильный. 
#Составьте закон распределения числа правильных ответов при простом угадывании. 
#Найдите M(X), D(X).

 #Проводится п независимых испытаний, в каждом из которых может произойти событие А.
#Тогда случайная величина х - число появления события А в данной сеотт испытаний имеет биномиальный закон распределения.
#Посчитаем для п = 5 и к = 0, 1, 2, 3, 4, 5. Вероятность появления р = 0.25, q = 1 - 0.25 = 0.75


from math import factorial

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
n = 5
p = 0.25
print('Закон распределения случайной величины:')
print('k    X')
for k in range(6):
    X = combinations(n, k) * (p ** k) * (1 - p) ** (n - k)
    print (k,'  ', X)
    
#Мат. ожидание и дисперсия:
print(f'Мат.ожидание случайной величины: {n * p}')
print(f'Дисперсия случайной величины: {n * p * (1 - p)}')
```

```python
# Задача 2
# Вероятность попадания в цель при одном выстреле равна 0.01. 
# Какова вероятность того, что число попаданий при 200 выстрелах будет не менее 5 и не более 10?

#При достаточно большом количестве испытаний п и малой вероятности р - распределение Пуассона приближается 

# к биномиальному. И lambda_ = n * p 
####

import numpy as np

def poisson_proba(k, lambda_):
    return (lambda_**k) * (np.exp(-lambda_))/ np.math.factorial(k)

n = 200
p = 0.01
lambda_ = n * p
X = 0

X = sum(poisson_proba(k=i, lambda_ = lambda_) for i in range (5,11))
print(f'Вероятность того, что число попаданий при 200 выстрелах будет не менее 5 и не более 10 составляет {X}')
```
