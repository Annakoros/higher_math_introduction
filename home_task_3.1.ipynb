```python
# 1. Написать код, реализующий расчет длины вектора, заданного его координатами
# Пусть вектор берет начало в начале координат (0, 0, 0)
coordinates_str = input('Введите координаты вектора через пробел: ')
coordinates = coordinates_str.split()
print(coordinates)
vector_length = 0
for coo in coordinates:
    vector_length += int(coo)**2
print(f'Длина вектора: {vector_length}')
```

```python
#Напишите код на Python, реализующий построение графиков: окружности, эллипса, гиперболы.
import numpy as np
import matplotlib.pyplot as plt
import decimal

R = 5
x = np.linspace(-10, 10, 41)
y = np.sqrt(R**2 - x**2)
y_ = -1*np.sqrt(R**2 - x**2)
plt.plot(x, y, color = 'b')
plt.plot(x, y_, color = 'b')

a = 7
b = 3
x = np.linspace(-10, 10, 41)
y = np.sqrt(b**2*(1 - x**2 / a**2))
y_ = -1*np.sqrt(b**2*(1 - x**2 / a**2))
plt.axis("equal")
plt.grid
plt.plot(x,y, color = 'g')
plt.plot(x,y_, color = 'g')

x = np.linspace(-10, 10, 41)
y = 1 / x
plt.plot(x, y, color = 'y')
plt.show()
```

```python
#Нарисуйте трехмерный график двух параллельных плоскостей.

from mpl_toolkits.mplot3d import Axes3D

xs = np.linspace(0, 10, 100)
zs = np.linspace(0, 10, 100)

X, Z = np.meshgrid(xs, zs)
Y = 5 - X
Y1 = 10 - X

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y1, Z)
plt.show()
```

```python
#Нарисуйте трехмерный график двух любых поверхностей второго порядка.
xs = np.linspace(-100, 100, 201)
ys = np.linspace(-100, 100, 201)

X, Y = np.meshgrid(xs, ys)
Z = 3*X**2 + 2*X*Y + 0.5*Y**2 + 5
Z1 = -0.3*X**2 - 0.5*X*Y - 5*Y**2 - 5

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z1)
plt.show()
```
