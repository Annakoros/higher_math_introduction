```python
import numpy as np

def arrangement(n, k):
    return np.math.factorial(n)//np.math.factorial(n-k)

def permutations(n):
    return np.math.factorial(n)

def combinations(n, k):
    return np.math.factorial(n)//(np.math.factorial(k)*np.math.factorial(n-k))

# Задача 1
#Группа студентов изучает 10 различных дисциплин. 
#Сколькими способами можно составить расписание на понедельник, 
#если в этот день должно быть 4 разных занятия?

#Порядок расположения предметов не важен, поэтому воспользуемся формулой для сочетаний:
c = combinations(10, 4)
print(c)
```

```python
#Задача 2
#Из колоды в 52 карты вынимают случайным образом 4 карты. 
#Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

#Число сочетаний при вытягивании 1 туза:
c_aces = combinations(4,1)
c_others = combinations(48,3)
a = c_aces * c_others
print(a)

#Число сочетаний при вытягивании 2 тузов:
c_aces = combinations(4,2)
c_others = combinations(48,2)
b = c_aces * c_others
print(b)

#Число сочетаний при вытягивании 3 тузов:
c_aces = combinations(4,3)
c_others = combinations(48,1)
c = c_aces * c_others
print(c)

#Число сочетаний при вытягивании 4 тузов:
c_aces = combinations(4,4)
c_others = combinations(48,0)
d = c_aces * c_others
print(d)

#Общее число сочетаний:
print(a + b + c + d)
```

```python
#Задача 3
#Семь человек рассаживаются наудачу на скамейке. 
#Какова вероятность того, что два определённых человека будут сидеть рядом?
#Число благоприятных исходов - случаи, когда два определенных человека располагаются рядом.
# Мысленно представляя их возможные места на скамейке, таких случаев: 6 + 6 = 12
#Общее же число исходов - количество перестанвок 7 человек:

perm_7 = permutations(7)

#Вероятность по классической формуле:

p_A = 12 / perm_7

print(p_A)
```

```python
#Задача 4
#Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. 
#Какова вероятность того, что среди трёх наугад выбранных вопросов 
#студент знает: а) все? б) два?

#Вычислим число исходов общее - сочетание 3 вопросов из 60:
n_general = combinations(60,3)

#а. Вычислим число комбинаций, когда студент вытянул билет с 3 вопросами из числа тех 50, что выучил
# и ни с одним из тех 10, что выучить не успел.
good_questions = combinations(50,3)
bad_questions = combinations(10,0)
a = good_questions * bad_questions

#Вычислим вероятность:
p_A = a / n_general
print(f'Вероятность того, что студент будет знать все вопросы в билете: {p_A}')

#б. Вычислим число комбинаций, когда студент вытянул билета с 2 вопросами из числа тех 50, что выучил
# и с 1 из тех 10, что выучить не успел.
good_questions = combinations(50,2)
bad_questions = combinations(10,1)
a = good_questions * bad_questions

#Вычислим вероятность:
p_A = a / n_general
print(f'Вероятность того, что студент будет знать 2 вопроса из 3: {p_A}')
```

```python
#Задача 6 (Дополнительно)
#Допустим, имеется некоторая очень редкая болезнь (поражает 0.1 % населения). 
#Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. 
#Врач говорит вам, что этот тест верно выявляет 99 % больных этой болезнью 
#и всего лишь в 1 % случаев даёт ложный положительный ответ.
#Вопрос: какова вероятность, что вы действительно больны ей?
#Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

#Событие А - положительный анализ. В1 - человек болен (вероятность 0.001), В2 - человек не болен (вероятность 0.999).
#Тогда полная вероятность положительного анализа:

p_positiv = 0.001 * 0.99 + 0.999 * 0.01

#Предположим, человек заболел. То есть наступило событие В1.
#Какова вероятность этого при положительном анализе:

p_real_sickness = (0.001 * 0.99)/ p_positiv

print(f'Вероятность того, что вы действительно заболели: {p_real_sickness}')
```
